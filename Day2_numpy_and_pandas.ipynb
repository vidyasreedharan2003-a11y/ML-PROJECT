```python
df=df.drop(columns=['deck'])
```

```python
df.shape
```

```python
df.dropna(subset=['embarked','embark_town'],inplace=True)  #To remove the null values inside the column
```

```python
print(df.isnull().sum())
```

```python
age_mean=df['age'].mean()
```

```python
df['age']=df['age'].fillna(age_mean)
```

```python
df['age_group']=np.where(df['age']<18,"Child","Adult")
```

```python
df['age_group']
```

```python
df['family_size']=df['sibsp']+df['parch']
```

```python
df['family_size']
```

```python
df['fare_per_pearson']=df['fare']/(df['family_size']+1)
```

```python
df['fare_per_pearson']
```

```python
df.rename(columns={'pclass':'passenger_class'},inplace=True)
```

```python
df
```

```python
df[df['age']>df['age'].mean()]   #To get age greater than its mean value
```

```python
df[(df['sex']=='female')&(df['passenger_class']==1)]  #to get female passengers with passenger class 1
```

```python
df[df['survived']==1]
```

```python
df.groupby('sex')['survived'].mean()   #survival rate by gender
```

```python
df.groupby('passenger_class')['fare'].mean()   #average fare by class
```

```python
df.corr(numeric_only=True)
```
